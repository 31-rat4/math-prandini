```python
######################################################
################ 1. Cogito Ergo Sum ##################
######################################################

import matplotlib.pyplot as plt
import numpy as np
import matplotlib.transforms as mtransforms
```

```python
#### Draw Graphikos for Cogito Ergo Sum

def draw_graphikos(name, func, init, end, desc_func):
    fig, ax = plt.subplots()

    ### Createing data
    x = [n for n in range(init, end)]
    y = [func(n) for n in x]
    zero = [0 for _ in x]

    ### Zero index
    ax.plot(x, zero, "black")
    ax.plot(zero, y, "black")


    ### Dots
    for nx, ny in zip(x,y):
        ax.text(nx,ny, f"({nx}, {ny})")

    ### Data
    ax.plot(x, y,color='tab:blue', marker='o')

    ### Headers
    ax.set_title(f"Dominus functus {name}\n{desc_func}")
    ax.set_xlabel('Dominus')
    ax.set_ylabel('Functus')
  
    return plt.show()
    
```

```python
##### f(x) = 2*x
def cogitoErgoSum_primus(dominus):
    functus = 2 * int(dominus)
    return functus
```

```python
draw_graphikos("Primus", cogitoErgoSum_primus, -4, 5, "f(x) = 2*x")
```

```python
#####f(x) = (x -2).(x-4)

def cogitoErgoSum_Secundus(dominus):
    functus = (int(dominus) -2)*(int(dominus) -4)
    return functus
```

```python
draw_graphikos("Secundus", cogitoErgoSum_Secundus, 0,7, "f(x) = (x -2).(x-4)")
```

```python
#####f(x) = (x -1).(x-3).(x-5)

def cogitoErgoSum_Tertium(dominus):
    functus = (int(dominus) -1)*(int(dominus) -3) *(int(dominus)-5)
    return functus
```

```python
draw_graphikos("Tertium", cogitoErgoSum_Tertium, 0,7, "f(x) = (x -1).(x-3).(x-5)")
```

```python
#####f(x) = (x -1).(x-3).(x-5).(x-7)

def cogitoErgoSum_Quartum(dominus):
    functus = (int(dominus) -1)*(int(dominus) -3) *(int(dominus)-5)*(int(dominus) - 7)
    return functus
```

```python
draw_graphikos("Quartum", cogitoErgoSum_Quartum, 0,9, "f(x) = (x -1).(x-3).(x-5).(x-7)")
```

```python
#####f(x) = (x -2).(x-2)

def cogitoErgoSum_Quintum(dominus):
    functus = (int(dominus) -2)*(int(dominus) -2)
    return functus
```

```python
draw_graphikos("Quintum", cogitoErgoSum_Quintum, 0,5, "f(x) = (x -2).(x-2)")
```

```python
#####f(x) = (x-1).(x -3).(x-3)

def cogitoErgoSum_Sextum(dominus):
    functus = (int(dominus)-1)*(int(dominus) -3)*(int(dominus) -3)
    return functus
```

```python
draw_graphikos("Sextum", cogitoErgoSum_Sextum, 0,5, "f(x) = (x-1).(x -3).(x-3)")
```

```python
###Zero division != True
def close_zero_division(dominus):
    return 1/dominus
```

```python
fig, ax = plt.subplots()
val_end = 6

### Createing data
n = 5
x = []
minus_x = []
for part in range(val_end):
    n /= 2 
    x.append(n)
    minus_x.append(n*-1)

y = [close_zero_division(n) for n in x]
minus_y = [close_zero_division(n) for n in minus_x]
zero = [0 for _ in x]

### Zero index
ax.plot(x, zero, "black")
ax.plot(zero, y, "black")
ax.plot(minus_x, zero, "black")
ax.plot(zero, minus_y, "black")


### Dots
for nx, ny in zip(x,y):
    ax.text(nx,ny, f"({nx:.2f}, {ny})")

trans_offset = mtransforms.offset_copy(ax.transData, fig=fig,
                                       x=-0.50, y=-0.20, units='inches')
for minus_nx, minus_ny in zip(minus_x, minus_y):
    ax.text(minus_nx, minus_ny, f"({minus_nx:.2f}, {minus_ny})", transform=trans_offset)

### Data
ax.plot(x, y,color='tab:blue', marker='o')
ax.plot(minus_x, minus_y,color='tab:blue', marker='o')

### Headers
ax.set_title("x ----> 0\n f(x)= 1/x -> infinity")
ax.set_xlabel('Dominus')
ax.set_ylabel('Functus')

plt.show()
```
